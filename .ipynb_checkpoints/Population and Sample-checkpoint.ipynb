```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Creating a column with 10 data = Dataframe

data = pd.DataFrame()
data['Population'] = [47, 48, 85, 20, 19, 13, 72, 16, 50, 60]
```

```python
# Drawing a sample replacement, size = 5, from population

sample_replacement = data['Population'].sample(5, replace=True)
print(sample_replacement)
```

```python
# without replacement: 
sample_without_replacement = data['Population'].sample(5, replace=False)
print(sample_without_replacement)
```

```python
population_mean = 43.0
population_var = 571.8
print('Population mean is ', population_mean)
print('Population variance is', population_var)
```

```python
# Calculate sample mean and sample standard deviation, size =10
# You will get different mean and varince every time when you excecute the below code

a_sample = data['Population'].sample(10, replace=True)
sample_mean = a_sample.mean()
sample_var = a_sample.var()
print('Sample mean is ', sample_mean)
print('Sample variance is', sample_var)
```

```python
sample_length = 500
sample_variance_collection=[data['Population'].sample(10, replace=True).var(ddof=1) for i in range(sample_length)]
```

```python
# Variation of a sample, where a sample mean and SD keeps changing but is always within a certain range:

Fstsample = pd.DataFrame(np.random.normal(10, 5, size=30))
print('sample mean is ', Fstsample[0].mean())
print('sample SD is ', Fstsample[0].std(ddof=1))
```

```python
meanlist = []
for t in range(10000):
    sample = pd.DataFrame(np.random.normal(10, 5, size=30))
    meanlist.append(sample[0].mean())
```

```python
collection = pd.DataFrame()
collection['meanlist'] = meanlist
```

```python
collection['meanlist'].hist(bins=100, density=True,figsize=(15,8))
```

```python
# sampling from an arbitrary distribution, which according to the central limit theorem
#, will give us a distribution of the sample mean to be approximately normal.

sample_size = 100
samplemeanlist = [] 
approx = pd.DataFrame([1, 0, 1, 0, 1])
for t in range(10000):
    sample = approx[0].sample(sample_size, replace=True) # small sample size
    samplemeanlist.append(sample.mean())

acollec = pd.DataFrame()
acollec['meanlist'] = samplemeanlist
acollec.hist(bins=100, density=True, figsize=(15,8))

```

```python
# Hypothesis testing: 

# 1. Import microsoft csv and add a new feature - logreturn: 

ms = pd.read_csv('MSFT.csv.xls')
ms['logReturn'] = np.log(ms['Close'].shift(-1)) - np.log(ms['Close'])
```

```python
# Log return goes up and down during the period: 
ms['logReturn'].plot(figsize=(20,8))
plt.axhline(0, color='red')
plt.show()
```
