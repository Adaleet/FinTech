```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("fmp-data.csv")  # Replace with the actual path to your file
```

```python
top_10_companies = df.head(10)

top_10_companies['stockExchange'].value_counts().plot(
    kind='bar', 
    title='Stock Exchange of Top 10 Companies',
    xlabel='Stock Exchange', 
    ylabel='Number of Companies'
)
plt.xticks(rotation=45, ha='right', rotation_mode='anchor')
plt.subplots_adjust(bottom=0.3)
plt.show()
```
